```python
#The levels in the pivot table will be stored in MultiIndex objects (hierarchical indexes) on the index and columns of the result DataFrame.
import pandas as pd
df = pd.read_csv("weather2.csv")
df
```

```python
# Pandas' pivot_table function operates similar to a spreadsheet, making it easier to group, summarize and analyze your data.
df.pivot_table(index='city',columns='date')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='sum')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='mean')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='mean',margins=True)
```
