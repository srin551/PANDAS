```python
#The merge() method updates the content of two DataFrame by merging them together, using the specified method(s).
import pandas as pd 
df1 = pd.DataFrame(
{
    "city": ['new york','chicago','orlando','us'],
    "temperature": [21,14,16,6]
})
df1
```

```python
df2 = pd.DataFrame(
{
    "city": ['new york','chicago','orlando',"canada"],
    "humidity": [22,15,6,8]
})
df2
```

```python
new_df = pd.merge(df1,df2,on="city")
```

```python
new_df
```

```python
#Database Join Functions
new_df = pd.merge(df1,df2,on="city",how = "inner")
new_df
```

```python
new_df = pd.merge(df1,df2,on="city",how = "inner")
new_df
```

```python
new_df = pd.merge(df1,df2,on="city",how = "left")
new_df
```

```python
new_df = pd.merge(df1,df2,on="city",how = "right")
new_df
```

```python
new_df = pd.merge(df1,df2,on="city",how = "outer")
new_df
```

```python
new_df = pd.merge(df1,df2,on="city",how = "outer",indicator=True)
new_df
```
