```python
#concat() in pandas works by combining Data Frames across rows or columns. We can concat two or more data frames either along rows (axis=0) or along columns (axis=1).
import pandas as pd
india_weather = pd.DataFrame({
    'city':['mumbai','delhi','haryan'],
    'humidity':[23,45,12],
    'temperature':[30,29,28]
    })
india_weather
```

```python
usa_weather = pd.DataFrame({
    'city':['chicago','newyork','canada'],
    'humidity':[23,2,2],
    'temperature':[13,19,28]
    
})
usa_weather
```

```python
df = pd.concat([usa_weather,india_weather])#performing concat operation 
df
```

```python
df = pd.concat([usa_weather,india_weather], ignore_index= True)
df
```

```python
df = pd.concat([usa_weather,india_weather],keys=["india","us"])#Using keys we can associate each key with each of the dataframe.
df
```
